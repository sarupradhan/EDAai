## Bivariate Analysis

```python
# Scatter plot(Numerical - Numerical) for the relation between two column 
sns.scatterplot(x='Age', y='Number of books you read during last 12months?',hue='Sex', style='Read Printed books lately?',data=data) 
plt.show()
```

```python
# Bar plot (Numerical - Categorical) 
# avverage age based on gender and Read Printed books lately?
sns.barplot(x='Sex',y='Age',hue='Read Printed books lately?',data=data)
```

```python
#Box plot (Numerical - Categorical) 
sns.boxplot(x='Sex',y='Age',hue='Read e-books lately?',data=data)
```

```python
# Displot (Distribution Plot)
sns.displot(data=data, x='Age', hue='Sex', kde=False, multiple="stack")
plt.show()
```

```python
# Displot (Distribution Plot)
sns.displot(data=data, x='Number of books you read during last 12months?', hue='Sex', kde=True, multiple="stack")
plt.show()
```

```python
# HeatMap (Categorical - Categorical) 
heatmap_data = pd.crosstab(data['Sex'], data['Race'])
sns.heatmap(heatmap_data, annot=True, cmap='PiYG', fmt='d')
```

```python
# pairplot 
sns.pairplot(data,hue='Sex')
```
